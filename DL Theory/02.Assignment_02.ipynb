### 1.	Describe the structure of an artificial neuron. How is it similar to a biological neuron? What are its main components?

**Biological Neurons**

Neurons are the basic functional units of the nervous system, and they generate electrical signals called action potentials, which allows them to quickly transmit information over long distances.
Almost all the neurons have three basic functions essential for the normal functioning of all the cells in the body.

These are to:
1. Receive signals (or information) from outside.
2. Process the incoming signals and determine whether or not the information should be passed along.
3. Communicate signals to target cells which might be other neurons or muscles or glands.

Now let us understand the basic parts of a neuron to get a deeper insight into how they actually work…

A biological neuron is mainly composed of 3 main parts and an external part called synapse:-

1. Dendrite

Dendrites are responsible for getting incoming signals from outside

2. Soma

Soma is the cell body responsible for the processing of input signals and deciding whether a neuron should fire an output signal

3. Axon

Axon is responsible for getting processed signals from neuron to relevant cells

4. Synapse

Synapse is the connection between an axon and other neuron dendrites

Working of the parts
The task of receiving the incoming information is done by dendrites, and processing generally takes place in the cell body. Incoming signals can be either excitatory — which means they tend to make the neuron fire (generate an electrical impulse) — or inhibitory — which means that they tend to keep the neuron from firing.

Most neurons receive many input signals throughout their dendritic trees. A single neuron may have more than one set of dendrites and may receive many thousands of input signals. Whether or not a neuron is excited into firing an impulse depends on the sum of all of the excitatory and inhibitory signals it receives. The processing of this information happens in soma which is neuron cell body. If the neuron does end up firing, the nerve impulse, or action potential, is conducted down the axon.

Towards its end, the axon splits up into many branches and develops bulbous swellings known as axon terminals (or nerve terminals). These axon terminals make connections on target cells.

Takes inputs from the input layer

Activation Function decides whether or not a neuron is fired. It decides which of the two output values should be generated by the neuron.

**Output Layer**
Output layer gives the final output of a neuron which can then be passed to other neurons in the network or taken as the final output value.

Initialize the weights: The algorithm starts by initializing the weights randomly. The number of weights is equal to the number of features in the data.

Update the weights: If the prediction is incorrect, the weights are updated in such a way as to reduce the prediction error. The update rule is given by the learning rule: w = w + (target - predicted) * x, where x is the input vector and target is the actual class.

Repeat steps 2 and 3: The algorithm continues to repeat steps 2 and 3 until the accuracy of the model reaches a satisfactory level or a maximum number of iterations is reached.

Finalize the model: After the iteration process has finished, the algorithm outputs the final weights, which can be used to classify new instances.

To classify a set of data using a simple perceptron, the algorithm is first trained on the data using the steps above, and then the final weights are used to predict the class of new instances.

### 4.	Explain the basic structure of a multi-layer perceptron. Explain how it can solve the XOR problem.

A multi-layer perceptron (MLP) is a type of artificial neural network that consists of multiple layers of artificial neurons. It is used for supervised learning and can perform non-linear classification and regression. The basic structure of an MLP is as follows:

Input Layer: This layer takes the input data and passes it to the hidden layer(s).

Hidden Layers: One or more hidden layers are present between the input and output layers. These layers are responsible for learning and transforming the input data into a higher-level representation that is easier for the output layer to use.

Output Layer: This layer takes the information from the hidden layer(s) and computes the final prediction.

Each artificial neuron in an MLP is connected to the neurons in the adjacent layer and uses weighted connections to propagate information. The weights are adjusted during the training process to minimize the error between the predicted and actual output.

The `XOR` problem refers to the problem of classifying a set of input data into two classes, where the output is 1 if the inputs belong to different classes and 0 if they belong to the same class. A single-layer perceptron is not capable of solving this problem because it can only separate the two classes with a linear boundary. However, an MLP with one hidden layer can solve the XOR problem by learning non-linear boundaries between the classes. This is achieved by using the activation function in the hidden layer, which allows the network to learn complex non-linear relationships between the inputs and outputs.

### 6.	Explain the learning process of an ANN. Explain, with example, the challenge in assigning synaptic weights for the interconnection between neurons? How can this challenge be addressed?

The learning process of an Artificial Neural Network (ANN) involves adjusting the synaptic weights of the connections between neurons in order to minimize the error between the predicted and actual outputs. The following is a general explanation of the learning process of an ANN:

Forward Propagation: In this phase, the input data is passed through the network, and the activations of each neuron are computed using the activation function and the weighted connections between neurons.

Computation of Error: The error between the predicted and actual outputs is calculated using a loss function, such as mean squared error or cross-entropy loss.

Backpropagation: In this phase, the error is propagated backwards through the network, and the weights are updated to minimize the error. This is done using an optimization algorithm such as gradient descent, which calculates the gradient of the loss with respect to the weights and updates the weights in the direction of steepest descent.

Assigning the correct synaptic weights is a challenge in the learning process of an ANN because it directly affects the performance of the network. If the weights are not assigned correctly, the network may not learn the correct relationships between the inputs and outputs, and the error may not decrease during the training process.

One approach to addressing this challenge is to use a weight initialization method that assigns the initial weights randomly, but in a way that ensures that the activations of the neurons are not too large or too small. This helps to prevent the problem of vanishing gradients, where the gradients become too small to update the weights effectively. Another approach is to use an optimization algorithm with a regularization term, such as L2 regularization, which penalizes large weights and helps to prevent overfitting.

Another approach is to use more advanced optimization algorithms, such as Adam or RProp, that adapt the learning rate for each weight based on the gradient of the loss with respect to that weight. This helps to ensure that the weights are updated in a way that converges to the optimal solution, even when the loss surface is complex.

### 7.	Explain, in details, the backpropagation algorithm. What are the limitations of this algorithm?

The backpropagation algorithm is a supervised learning algorithm that is used to train Artificial Neural Networks (ANNs). It is an iterative algorithm that adjusts the weights of the connections between neurons in order to minimize the error between the predicted and actual outputs. The following is a detailed explanation of the backpropagation algorithm:

Forward Propagation: The input data is passed through the network, and the activations of each neuron are computed using the activation function and the weighted connections between neurons. The activations are stored in a data structure called the activation cache.

Computation of Error: The error between the predicted and actual outputs is calculated using a loss function, such as mean squared error or cross-entropy loss.

Backward Propagation: In this phase, the error is propagated backwards through the network, and the gradient of the loss with respect to each weight is computed using the chain rule of calculus. This is done by computing the derivative of the loss with respect to the activations of each neuron, and using the stored activation cache from the forward propagation phase to compute the gradient with respect to each weight.

Update of Weights: The weights are updated using an optimization algorithm, such as gradient descent, which calculates the gradient of the loss with respect to the weights and updates the weights in the direction of steepest descent. The learning rate is a hyperparameter that determines the size of the update, and it can be tuned to achieve the best performance.

The backpropagation algorithm is widely used because it is computationally efficient, and it can be used to train deep neural networks with multiple hidden layers. However, it has some limitations that should be considered:

Local Minima: The backpropagation algorithm is based on gradient descent, which is susceptible to getting stuck in local minima, where the error is lower than the global minimum, but not the lowest possible error.

Vanishing Gradients: The backpropagation algorithm can suffer from the problem of vanishing gradients, where the gradients become too small to update the weights effectively. This can occur when the activation functions have a small slope, which makes it difficult for the network to learn complex relationships between the inputs and outputs.

Overfitting: The backpropagation algorithm can also lead to overfitting, where the network fits the training data too closely and fails to generalize to new data. This can be addressed using regularization techniques, such as L2 regularization, or by using a smaller network.

In summary, the backpropagation algorithm is an effective and efficient way to train ANNs, but it has some limitations that should be considered and addressed in order to achieve the best performance.

### 8.	Describe, in details, the process of adjusting the interconnection weights in a multi-layer neural network.

The process of adjusting the interconnection weights in a multi-layer neural network is known as training. The goal of training is to find the optimal values for the interconnection weights that minimize the difference between the network's predicted outputs and the actual outputs. The training process consists of the following steps:

Forward Propagation: The input data is fed into the input layer of the network, and then passed forward through the hidden layers to the output layer. At each layer, the weighted sum of inputs is computed and passed through an activation function to produce the output of that neuron.

Loss Function: The difference between the network's predicted outputs and the actual outputs is measured using a loss function. The loss function is a measure of how well the network is performing, and is used to guide the optimization process.

Back Propagation: The gradient of the loss function with respect to the interconnection weights is calculated using the backpropagation algorithm. The gradient indicates the direction in which the weights need to be adjusted to reduce the loss.

Weight Update: The interconnection weights are updated in the direction of the negative gradient, using an optimization algorithm such as gradient descent. The step size of the update is controlled by the learning rate, which is a hyperparameter that determines the size of the update.

Repeat: The process of forward propagation, loss calculation, backpropagation, and weight update is repeated for multiple epochs, until the loss converges to a minimum or reaches a satisfactory level.

This process of adjusting the interconnection weights is an iterative process, and it can take many epochs to reach a satisfactory solution. The optimal values of the weights depend on the specific task and data, and can be influenced by factors such as the architecture of the network, the choice of activation functions, the learning rate, and the choice of optimization algorithm.

**Activation Function vs Threshold Function:**
An activation function is a non-linear mathematical function applied to the weighted sum of inputs in an artificial neuron, determining the output of the neuron. Activation functions introduce non-linearity into the model, allowing it to capture complex patterns in the data.

**A threshold function** is a simple activation function that outputs either 0 or 1, based on a threshold value. This threshold value is typically set such that the output of the neuron is 1 if the weighted sum of inputs is greater than or equal to the threshold, and 0 otherwise. Threshold functions are less commonly used in modern neural networks, as they are less flexible and cannot capture complex patterns in the data compared to more advanced activation functions such as sigmoid or ReLU.

**Step Function vs Sigmoid Function:**
A step function is a simple activation function that outputs either 0 or 1, based on a threshold value. This threshold value is typically set such that the output of the neuron is 1 if the weighted sum of inputs is greater than or equal to the threshold, and 0 otherwise.

**A sigmoid function** is a more sophisticated activation function that maps any input value to a value between 0 and 1. The output of the sigmoid function approaches 1 as the input increases, and approaches 0 as the input decreases. The sigmoid function is commonly used in the output layer of binary classification problems, where it allows the model to make probabilistic predictions.

**Single Layer vs Multi-layer Perceptron:**
A single layer perceptron is a type of artificial neural network that consists of a single layer of artificial neurons. It is a simple model that can only solve linearly separable problems, and is limited in its ability to capture complex patterns in the data.

**A multi-layer perceptron (MLP)** is a type of artificial neural network that consists of multiple layers of artificial neurons, including at least one hidden layer. The hidden layers perform non-linear transformations to capture complex patterns in the data, making MLPs more powerful and flexible than single layer perceptrons. MLPs are widely used for pattern recognition tasks, such as image and speech recognition, and have proven to be effective in solving non-linear problems.
